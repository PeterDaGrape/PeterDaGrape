```python
from naughtscrosses import Game 
import numpy as np
import random
```

```python
def flatten(state):
    return np.array(state).reshape(-1)
```

```python
class Model:
    def __init__(self, table_shape, num_actions):

        self.epsilon = 0.6
        self.epsilonDecay = 0.999
        self.discount = 0.95
        self.learning_rate = 0.1
        self.reward = 0
        self.q_table = np.random.uniform(low = -2, high = 0, size = (table_shape + [num_actions]))


    def predict(self, state):

        
        action = np.argmax(self.q_table[state])
        print(self.q_table[state])



        return action    
    

    def update_table(self, state, new_state, action):
        
        max_future_q = np.max(self.q_table[new_state])

        current_q = self.q_table[state + (action, )]

        new_q = (1 - self.learning_rate) * current_q + self.learning_rate * (self.reward + self.discount * max_future_q)

        self.q_table[state + (action, )] = new_q


        self.reward = 0
```

```python
num_epsisodes = 2000

model_1 = Model([9, ], 8)
model_2 = Model([9, ], 8)
```

```python

for episode in range(num_epsisodes):
    game = Game()


    move = 0


    while not game.game_over:
        if move % 2 == 0:
            model = model_1
            player = 1
        else:
            model = model_2
            player = 2

        state = flatten(game.state)
        
        
        action = model.predict(state)
        print(action)

        while True:

            x = int(action / 3)
            y = action % 3

            if game.update_game(x, y, player):

                break
            else:
                action = random.randint(0, 8)



        winner = game.check_winner()

        if winner == 1:
            model_1.reward += 2
        if winner == 2:
            model_2.reward += 2


        model.update_table(state, flatten(game.state), action)


        game.display_game()

        move += 1


```
